```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("hr_comma_sep.csv")
df.head()
```

```python
left = df[df['left'] == 1]
retained = df[df['left'] == 0]
```

```python
df.groupby('left').mean(numeric_only = True)
```

```python
# The numeric factors that impact whether an employee leaves a firm are: satisfaction_level, average_monthly_hours, and promotion_last_5years
```

```python
newdf = df[['satisfaction_level', 'average_montly_hours', 'promotion_last_5years', 'salary']]
```

```python
dummies = pd.get_dummies(df['salary'], dtype = int)
dummies
```

```python
merged = pd.concat([newdf, dummies], axis = 'columns') 
merged.head()
```

```python
finaldf = merged.drop(['salary'], axis = 'columns')
```

```python
finaldf.head()
```

```python
X = finaldf
y = df['left']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lreg = LogisticRegression()
```

```python
lreg.fit(X_train, y_train)
```

```python
lreg.predict(X_test)
```

```python
lreg.score(X_test, y_test)
```
