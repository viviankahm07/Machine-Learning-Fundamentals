```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target == 2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], marker = '+', color = 'green')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], marker = '+', color = 'blue')
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], marker = '+', color = 'green')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], marker = '+', color = 'blue')
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['target', 'flower_name'], axis = 'columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
